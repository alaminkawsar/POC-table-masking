```python
import cv2
import pytesseract

img = cv2.imread("data/images/ss-1.jpeg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python

data = pytesseract.image_to_data(
    gray,
    lang="eng",
    output_type=pytesseract.Output.DICT
)

n_boxes = len(data["text"])

for i in range(n_boxes):
    text = data["text"][i]
    conf = int(data["conf"][i])

    if conf > 20 and text.strip() != "":
        x = data["left"][i]
        y = data["top"][i]
        w = data["width"][i]
        h = data["height"][i]

        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(img, text, (x, y - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)

cv2.imwrite("Text Boxes.png", img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
# gray image show
cv2.imshow("Gray Image", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# make binary image
_, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)
cv2.imshow("Binary Image", binary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# now apply OCR on binary image
data_binary = pytesseract.image_to_data(
    binary,
    lang="eng",
    output_type=pytesseract.Output.DICT
)
n_boxes_binary = len(data_binary["text"])
for i in range(n_boxes_binary):
    text = data_binary["text"][i]
    conf = int(data_binary["conf"][i])

    if conf > 20 and text.strip() != "":
        x = data_binary["left"][i]
        y = data_binary["top"][i]
        w = data_binary["width"][i]
        h = data_binary["height"][i]
        print(text, conf, x, y, w, h)
        cv2.rectangle(binary, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(binary, text, (x, y - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)
cv2.imwrite("Text Boxes on Binary Image.png", binary)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
